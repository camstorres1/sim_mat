### 1. Graficar las siguientes funciones f(x). Determina el dominio a partir de la función:

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0, 20, 100)
y = x**2 - 4*x +2
y2 = np.sin(3*x)
y3 = math.e**(2*x) - 3
y4 = (x**2 - 4)**0.5
y5 = np.cos(x + (math.pi/4))

plt.figure(figsize=(8,2))
plt.plot(x, y, "r.", label = "primera linea")
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Título")
plt.plot(x, y2, "g*", label = "segunda linea")
plt.plot(x, y3, "b", label = "tercera linea")
plt.legend(loc="best")
```

### 3. Crea una función que cree una gráfica y dibuje 'k' líneas de formaimage.png. La función recibe:

```python
import matplotlib.pyplot as plt
import numpy as np

def lineas(x1, x2, n, a, b, k):
    x = np.linspace(x1, x2, n)
    va = np.linspace(a, b, k)

    plt.figure(figsize=(8,2))
    for i, ai in enumerate(va):
        y = np.sin(x*ai)
        plt.plot(x, y, "r.", label = f'a = {ai:2f}')
    plt.grid()
    plt.xlabel("Eje x")
    plt.ylabel("Eje y")
    plt.title("Título")
    plt.legend()
    plt.show()
lineas(x1=0, x2=2*np.pi, n=1000, a=1, b=3, k=3)
```
